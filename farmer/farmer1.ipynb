```python
import pandas as pd
import  matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/upflairs-pvt-ltd/3rd_july_datascience/refs/heads/master/farmer_guider/farmer.csv")
df.head()
```

```python
df["label"].value_counts()
```

```python
df.info()
```

```python
# unsupervised learning:
# this problem is the clustering analysis
# grouping
# K means algorithmm will be used
# in unsupervised learning there is only x
group=df.groupby("label")
```

```python
df["temperature"].min()
```

```python
group[["temperature"]].min()
```

```python
group[["temperature"]].mean()
```

```python
tem_sampleanalysis=group[["temperature"]].agg(["min","max","mean"])  # aaggregate function
```

```python
tem_sampleanalysis.reset_index(inplace=True)
tem_sampleanalysis
```

```python

plt.bar(tem_sampleanalysis["label"], tem_sampleanalysis["temperature"]["min"])
plt.xlabel("Crops")
plt.ylabel("Minimum Temperature")
plt.title("Minimum Temperature for Different Crops")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.bar(tem_sampleanalysis["label"], tem_sampleanalysis["temperature"]["max"])
plt.xlabel("Crops")
plt.ylabel("Maximum Temperature")
plt.title("Maximum Temperature for Different Crops")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.bar(tem_sampleanalysis["label"], tem_sampleanalysis["temperature"]["mean"])
plt.xlabel("Crops")
plt.ylabel("average Temperature")
plt.title("Average Temperature for Different Crops")
plt.xticks(rotation=90)
plt.show()
```

```python
df2=df.drop("label",axis=1)
```

```python
df2.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
std_scaler=StandardScaler()
```

```python
std_scaler.fit(df2)
```

```python
x=std_scaler.transform(df2)
```

```python
x
```

```python
x.shape
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans=KMeans()
```

```python
kmeans.fit(x)
```

```python
cluster_no=kmeans.labels_
cluster_no
```

```python
cluster_no.shape
```

```python
df
```

```python
df["cluster_no"]=cluster_no
```

```python
df.head()
```

```python
df["cluster_no"].unique()
```

```python
cluster_1=df[df["cluster_no"]==1]
```

```python
cluster_1["label"].value_counts()
```

```python
list(cluster_1["label"].unique())
```

```python
import os,joblib
```

```python
os.makedirs("model",exist_ok=True)
joblib.dump(kmeans,"model/kmeans.lb")
joblib.dump(std_scaler,"model/std.lb")
```
